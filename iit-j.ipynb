```python
import matplotlib.pyplot as plt
import numpy  as np
x_data=np.array([1,2,3,4,5,6,7,8])
y_data=np.array([2.1,3.9,6.2,7.8,10.1,12.3,14.0,15.8])

#creating a model line(linear fit)
coeffs=np.polyfit(x_data,y_data,1)
x_line=np.linspace(1,8,100)
y_line=coeffs[0]*x_line+coeffs[1]

#create plot
plt.figure(figsize=(8,6))
plt.scatter(x_data,y_data,color='blue',s=80,alpha=0.7,label='Data Points')
plt.plot(x_line,y_line,color='green',linewidth=2,label="Model Line")

#Add labels and formatting
plt.xlabel("xvalues",fontsize=12)
plt.ylabel("yvalues",fontsize=12)
plt.title("Scatter Plot with Linear Model",fontsize=14)
plt.legend()
plt.grid(True,alpha=0.3)
plt.show()
```
